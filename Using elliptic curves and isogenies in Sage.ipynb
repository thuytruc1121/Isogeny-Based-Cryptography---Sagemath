# Elliptic Curves

In Sage, an elliptic curve is always specified by the coeffs of a long Weirestrass equation
y^2 + a_1xy + a_3y = x^3 + a_2x^2 + a_4x + a_6

```python
Ex: a_1 = 1, a_2 = 3, a_3 =2, a_4 = 4,a_6 = 5
```

```python
E = EllipticCurve([1,3,2,4,5]);E
```

```python
Short Weirestrass equation
y^2 = x^3 + ax + b
```

```python
E = EllipticCurve([0,0,0,4,5]);E
```

```python
E = EllipticCurve([4,5]);E
```
